```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
temperatura = np.array([20, 23, 24, 25, 26]).reshape(-1, 1)
cantidad_piratas = np.array([50, 40, 30, 20, 10]) 

modeloPoly = PolynomialFeatures(degree=2)

temperatura_poly = modeloPoly.fit_transform(temperatura)

modelo_ln = LinearRegression()

modelo_ln.fit(temperatura_poly, cantidad_piratas)

prediccion_piratas_poly = modelo_ln.predict(temperatura_poly)

print ("score", modelo_ln.score(temperatura_poly,cantidad_piratas))
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(cantidad_piratas, temperatura, color='blue', label='Datos')
plt.plot(prediccion_piratas_poly, temperatura, color='red', label='Regresión polinomial')

plt.title('Correlación entre temperatura y cantidad de piratas (Regresión Polinomial)')
plt.xlabel('Cantidad de piratas')
plt.ylabel('Temperatura (°C)')

plt.legend()

plt.show()
```

```python
temperatura = np.array([20, 23, 24, 25, 26])
cantidad_piratas = np.array([50, 40, 30, 20, 10]) 

coeficientes = np.polyfit(cantidad_piratas, temperatura, 2)
polinomio = np.poly1d(coeficientes)
cantidad_piratas_curva = np.linspace(cantidad_piratas.min(), cantidad_piratas.max(), 100)
temperatura_curva= polinomio(cantidad_piratas_curva)

plt.figure(figsize=(6, 4))
plt.scatter(cantidad_piratas, temperatura, color='blue', label='Datos')
plt.plot(cantidad_piratas_curva, temperatura_curva, color='red', label='Regresión polinomial')

plt.title('Correlación entre temperatura y cantidad de piratas (Regresión Polinomial)')
plt.xlabel('Cantidad de piratas')
plt.ylabel('Temperatura (°C)')

plt.legend()
plt.show()
```

```python
temperatura = np.array([20, 23, 24, 25, 26])
cantidad_piratas = np.array([50, 40, 30, 20, 10]).reshape(-1, 1)

modeloPoly = PolynomialFeatures(degree=3)

piratas_poly = modeloPoly.fit_transform(cantidad_piratas)

modelo_ln = LinearRegression()

modelo_ln.fit(piratas_poly, temperatura)

prediccion_temp_poly = modelo_ln.predict(piratas_poly)

print ("score", modelo_ln.score(piratas_poly,temperatura))

coeficientes = modelo_ln.coef_
intercepto = modelo_ln.intercept_
polinomio = np.poly1d([coeficientes[0], coeficientes[1], intercepto])

cantidad_piratas_curva = np.linspace(cantidad_piratas.min(), cantidad_piratas.max()).reshape(-1, 1)
piratas_poly_curva = modeloPoly.transform(cantidad_piratas_curva)
temperatura_curva= modelo_ln.predict(piratas_poly_curva)

plt.figure(figsize=(6, 4))
plt.scatter(cantidad_piratas, temperatura, color='blue', label='Datos')
plt.plot(cantidad_piratas_curva, temperatura_curva, color='red', label='Regresión polinomial')

plt.title('Correlación entre temperatura y cantidad de piratas (Regresión Polinomial)')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Cantidad de piratas')

plt.legend()
plt.gca().invert_xaxis()
plt.show()
```
